<h1>First Webscrape Attempt</h1>
<p>Got it scraped, but didn't get the anchor text to accompany the href links</p>

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape
url = 'https://www.supremecourt.ohio.gov/courts/judicial-system/ohio-trial-courts/'

# Send a GET request to the webpage
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all elements with "aria-controls" attribute
elements = soup.find_all(attrs={"aria-controls": True})

# Iterate over the elements
for element in elements:
    # Get the name from "aria-controls" attribute
    name = element['aria-controls']
    print(f"Name: {name}")

    # Find the corresponding <div> element based on "aria-controls" attribute value
    div_id = f"collapse_{name}"
    div_element = soup.find(id=div_id)

    # Check if the div_element exists
    if div_element:
        # Find all <a> tags within the parent of the <div> element
        links = div_element.find_all('a')

        # Iterate over the links and extract the href attribute
        for link in links:
            href = link.get('href')
            print(f"Link: {href}")
    else:
        print("No corresponding <div> element found.")

    print()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = 'https://www.supremecourt.ohio.gov/courts/judicial-system/ohio-trial-courts/'

# Send a GET request to the webpage
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all elements with "aria-controls" attribute
elements = soup.find_all(attrs={"aria-controls": True})

# Initialize lists to store the data
names = []
links = []

# Iterate over the elements
for element in elements:
    # Get the name from "aria-controls" attribute
    name = element['aria-controls']
    names.append(name)

    # Find the corresponding <div> element based on "aria-controls" attribute value
    div_id = f"collapse_{name}"
    div_element = soup.find(id=div_id)

    # Check if the div_element exists
    if div_element:
        # Find all <a> tags within the parent of the <div> element
        anchor_tags = div_element.find_all('a')

        # Extract the href attribute from the anchor tags
        hrefs = [tag.get('href') for tag in anchor_tags]
        links.append(hrefs)
    else:
        links.append([])

# Create a DataFrame from the collected data
data = {'Name': names, 'Links': links}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

<h1>Third Webscrape Attempt</h1>
<p>Managed to create a spot in the tuple for the anchor text, but the code did not correctly read the anchor text or store it into our data. Almost there!</p>

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = 'https://www.supremecourt.ohio.gov/courts/judicial-system/ohio-trial-courts/'

# Send a GET request to the webpage
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all elements with "aria-controls" attribute
elements = soup.find_all(attrs={"aria-controls": True})

# Initialize lists to store the data
names = []
links = []
titles = []

# Iterate over the elements
for element in elements:
    # Get the name from "aria-controls" attribute
    name = element['aria-controls']
    names.append(name)

    # Find the corresponding <div> element based on "aria-controls" attribute value
    div_id = f"collapse_{name}"
    div_element = soup.find(id=div_id)

    # Check if the div_element exists
    if div_element:
        # Find all <a> tags within the parent of the <div> element
        anchor_tags = div_element.find_all('a')

        # Extract the href and title attributes from the anchor tags
        hrefs = [tag.get('href') for tag in anchor_tags]
        titles = [tag.get('title') for tag in anchor_tags]

        links.append(list(zip(hrefs, titles)))
    else:
        links.append([])

# Create a DataFrame from the collected data
data = {'Name': names, 'Links': links}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

<h1>Fourth Time's A Charm</h1>
<p>Finally got the anchor text to accompany the href links!</p>

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = 'https://www.supremecourt.ohio.gov/courts/judicial-system/ohio-trial-courts/'

# Send a GET request to the webpage
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all elements with "aria-controls" attribute
elements = soup.find_all(attrs={"aria-controls": True})

# Initialize lists to store the data
names = []
links = []

# Iterate over the elements
for element in elements:
    # Get the name from "aria-controls" attribute
    name = element['aria-controls']
    names.append(name)

    # Find the corresponding <div> element based on "aria-controls" attribute value
    div_id = f"collapse_{name}"
    div_element = soup.find(id=div_id)

    # Check if the div_element exists
    if div_element:
        # Find all <a> tags within the parent of the <div> element
        anchor_tags = div_element.find_all('a')

        # Extract the href and anchor text from the anchor tags
        links_with_text = [(tag['href'], tag.text.strip()) for tag in anchor_tags]

        links.append(links_with_text)
    else:
        links.append([])

# Create a DataFrame from the collected data
data = {'Name': names, 'Links': links}
ohio_df = pd.DataFrame(data)

# Display the DataFrame
print(ohio_df)
```

```python
#Import to csv
ohio_df.to_csv('Resources/ohio_df.csv', index=False)
```
